#**Theoretical**

**1. Can we use Bagging for regression problems?**

  Bagging (Bootstrap Aggregating) can definitely be used for regression problems.

**2 What is the difference between multiple model training and single model training?**

  Single model training is simpler, faster, and easier to understand, Train one model on the entire dataset,Learn one function to map inputs to output,Linear Regression, Decision Tree, SVM,Uses only one model's prediction.
  Multiple model training (ensemble methods) is more powerful and accurate, especially on complex datasets, but requires more computation and can be harder to interpret,Train many models, combine their outputs.

**3. Explain the concept of feature randomness in Random Forest?**

  In a Random Forest, feature randomness refers to the random selection of a subset of features (columns) at each split of every decision tree.
  This is different from a regular decision tree, which tries all features at each split to find the best one.

**4. What is OOB (Out-of-Bag) Score?**

  OOB (Out-of-Bag) Score is a built-in validation method used in ensemble models like Random Forests that use bagging (Bootstrap Aggregating).
  It provides an estimate of model accuracy without needing a separate validation set or cross-validation.

**5. How can you measure the importance of features in a Random Forest model?**

  In a Random Forest model, measuring feature importance helps to understand which features contribute the most to the prediction.
  Random Forest estimates feature importance in two main ways:
  1.Mean Decrease in Impurity (MDI) - Default method
  Also known as Gini Importance (for classification) or Variance Reduction (for regression):
  Every time a feature is used to split a node, the model computes how much the impurity (like Gini impurity or MSE) is reduced.
  These reductions are averaged over all trees in the forest.
  The more a feature reduces impurity, the more important it is.
  This is what's typically returned by .feature_importances_ in scikit-learn.
  2.Permutation Importance (Mean Decrease in Accuracy) - Model-agnostic
  Randomly shuffle (permute) the values of a feature in the dataset.
  Measure the drop in model accuracy (or increase in error).
  A bigger drop means the feature was important.
  This method is more reliable and robust, especially when features are correlated.

**6. Explain the working principle of a Bagging Classifier?**

  Working Principle of a Bagging Classifier:
  1.Bootstrapping the Data: From the original training dataset, create multiple random subsets by sampling with replacement.
  Each subset is called a bootstrap sample.
  Each bootstrap sample is typically the same size as the original dataset, but due to sampling with replacement, some observations may repeat and some may be left out (called OOB samples).
  2.Train Base Models: For each bootstrap sample, train a base classifier (often a Decision Tree).
  All classifiers are trained independently.
  3.Aggregation (Voting): For classification, predictions from all base models are collected and the majority vote is taken as the final prediction.This reduces the variance of the prediction, making it more stable and robust.

**21. Train a Bagging Classifier using Decision Trees on a sample dataset and print model accuracy?**


```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Load the sample dataset (Iris)
X, y = load_iris(return_X_y=True)

# 2. Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Create a base estimator (Decision Tree)
base_model = DecisionTreeClassifier(random_state=42)

# 4. Create the Bagging Classifier
bagging_clf = BaggingClassifier(
    estimator=base_model,
    n_estimators=50,            # number of base models
    bootstrap=True,             # use bootstrapping
    random_state=42
)

# 5. Train the Bagging Classifier
bagging_clf.fit(X_train, y_train)

# 6. Make predictions
y_pred = bagging_clf.predict(X_test)

# 7. Evaluate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Bagging Classifier Accuracy:", accuracy)
```

**27.Train a Random Forest Classifier with different numbers of trees and compare accuracy.**


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load dataset
X, y = load_breast_cancer(return_X_y=True)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Try different numbers of trees
tree_counts = [10, 50, 100, 200]
accuracies = []

# Train and evaluate each model
for n in tree_counts:
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"Accuracy with {n} trees: {acc:.4f}")

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(tree_counts, accuracies, marker='o', linestyle='--')
plt.title("Random Forest Accuracy vs Number of Trees")
plt.xlabel("Number of Trees (n_estimators)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

**31. Train a Random Forest Classifier and tune hyperparameters using GridSearchCV.**


```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

# 1. Load dataset
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Define the model
rf = RandomForestClassifier(random_state=42)

# 3. Define parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}

# 4. Setup GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# 5. Train the grid search
grid_search.fit(X_train, y_train)

# 6. Evaluate the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# 7. Print results
print("Best Parameters:", grid_search.best_params_)
print("Test Set Accuracy:", accuracy)
```

**32.Train a Bagging Regressor with different numbers of base estimators and compare performance.**


```python
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# 1. Load the dataset
X, y = fetch_california_housing(return_X_y=True)

# 2. Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Define estimator counts to compare
estimator_counts = [10, 50, 100, 200]
mse_scores = []

# 4. Train and evaluate BaggingRegressor with different n_estimators
for n in estimator_counts:
    model = BaggingRegressor(
        estimator=DecisionTreeRegressor(),
        n_estimators=n,
        random_state=42,
        n_jobs=-1
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)
    print(f"MSE with {n} estimators: {mse:.4f}")

# 5. Plot MSE vs. Number of Estimators
plt.figure(figsize=(8, 5))
plt.plot(estimator_counts, mse_scores, marker='o', linestyle='--', color='darkblue')
plt.title("Bagging Regressor Performance vs Number of Estimators")
plt.xlabel("Number of Base Estimators")
plt.ylabel("Mean Squared Error (MSE)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load Dataset
iris = load_iris()
X, y = iris.data, iris.target

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Define Base Learners
clf_tree = DecisionTreeClassifier(random_state=42)
clf_svm = SVC(probability=True, random_state=42)
clf_log = LogisticRegression(max_iter=1000, random_state=42)

# Step 4: Define Stacking Classifier
estimators = [
    ('dt', clf_tree),
    ('svm', clf_svm),
    ('log', clf_log)
]

stacking = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5
)

# Step 5: Train & Predict Models Individually
clf_tree.fit(X_train, y_train)
y_tree = clf_tree.predict(X_test)

clf_svm.fit(X_train, y_train)
y_svm = clf_svm.predict(X_test)

clf_log.fit(X_train, y_train)
y_log = clf_log.predict(X_test)

stacking.fit(X_train, y_train)
y_stack = stacking.predict(X_test)

# Step 6: Accuracy Comparison
print(f"Decision Tree Accuracy: {accuracy_score(y_test, y_tree):.2f}")
print(f"SVM Accuracy: {accuracy_score(y_test, y_svm):.2f}")
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_log):.2f}")
print(f"Stacking Classifier Accuracy: {accuracy_score(y_test, y_stack):.2f}")
```

**39.Train a Random Forest Classifier and analyze the effect of max_depth on accuracy.**


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load Dataset
iris = load_iris()
X, y = iris.data, iris.target

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train Random Forest with different max_depth values
depths = range(1, 21)
accuracies = []

for depth in depths:
    rf = RandomForestClassifier(n_estimators=100, max_depth=depth, random_state=42)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

# Step 4: Plot max_depth vs accuracy
plt.figure(figsize=(8, 5))
plt.plot(depths, accuracies, marker='o', linestyle='-', color='teal')
plt.title("Effect of max_depth on Random Forest Accuracy")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

**40. Train a Bagging Regressor using different base estimators (DecisionTree and KNeighbors) and compare performance.**

**41.Train a Random Forest Classifier and evaluate its performance using ROC-AUC Score.**


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Step 1: Load Dataset
iris = load_iris()
X, y = iris.data, iris.target
y_bin = label_binarize(y, classes=[0, 1, 2])  # One-hot encoding

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

# Step 3: Train Random Forest Classifier
rf = OneVsRestClassifier(RandomForestClassifier(n_estimators=100, random_state=42))
rf.fit(X_train, y_train)

# Step 4: Predict Probabilities
y_score = rf.predict_proba(X_test)

# Step 5: ROC-AUC Score
roc_auc = roc_auc_score(y_test, y_score, average="macro")
print(f"ROC-AUC Score (Macro Average): {roc_auc:.3f}")

# Step 6: Plot ROC Curves
fpr = {}
tpr = {}
for i in range(y_bin.shape[1]):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])

plt.figure(figsize=(7,5))
for i in range(y_bin.shape[1]):
    plt.plot(fpr[i], tpr[i], label=f"Class {i} ROC curve")

plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curves - Random Forest Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```

**44. Train a Stacking Classifier with Random Forest and Logistic Regression and compare accuracy.**


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Step 1: Load Dataset
iris = load_iris()
X, y = iris.data, iris.target

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Base Models
rf = RandomForestClassifier(n_estimators=100, random_state=42)
log = LogisticRegression(max_iter=1000, random_state=42)

# Step 4: Stacking Classifier (Meta Learner = Logistic Regression)
stacking = StackingClassifier(
    estimators=[('rf', rf), ('log', log)],
    final_estimator=LogisticRegression(),
    cv=5
)

# Step 5: Train Models Individually
rf.fit(X_train, y_train)
log.fit(X_train, y_train)
stacking.fit(X_train, y_train)

# Step 6: Predictions
y_pred_rf = rf.predict(X_test)
y_pred_log = log.predict(X_test)
y_pred_stack = stacking.predict(X_test)

# Step 7: Accuracy Comparison
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}")
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred_log):.2f}")
print(f"Stacking Classifier Accuracy: {accuracy_score(y_test, y_pred_stack):.2f}")
```

**45.Train a Bagging Regressor with different levels of bootstrap samples and compare performance.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load Dataset
data = fetch_california_housing()
X, y = data.data, data.target

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Define different bootstrap sample sizes
sample_fractions = [0.4, 0.6, 0.8, 1.0]
mse_scores = []
r2_scores = []

# Step 4: Train Bagging Regressor with different max_samples
for frac in sample_fractions:
    model = BaggingRegressor(
        estimator=DecisionTreeRegressor(),
        n_estimators=50,
        max_samples=frac,
        random_state=42
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    mse_scores.append(mse)
    r2_scores.append(r2)

    print(f"Bootstrap Fraction: {frac} -> MSE: {mse:.3f}, R²: {r2:.3f}")

# Step 5: Plot Performance vs Bootstrap Fraction
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(sample_fractions, mse_scores, marker='o', color='red')
plt.title("MSE vs Bootstrap Sample Size")
plt.xlabel("Bootstrap Fraction (max_samples)")
plt.ylabel("Mean Squared Error")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(sample_fractions, r2_scores, marker='o', color='blue')
plt.title("R² Score vs Bootstrap Sample Size")
plt.xlabel("Bootstrap Fraction (max_samples)")
plt.ylabel("R² Score")
plt.grid(True)

plt.tight_layout()
plt.show()
```
